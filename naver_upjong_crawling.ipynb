#### 누군가는 쓸 일이 있겠지 라는 생각으로 코랩 환경에서 작성했다.<br><br>
#### 개인적으로 크레온 업종 코드보다는 네이버 업종 코드가 최신이고 분류도 더 잘된 것 같아서 직접 크롤링 했다.<br><br>

#### 그러나 최근에 내가 보고 있던 종목의 회사명이 바뀐 적이 있었다.<br><br>
#### 아무것도 모르고 회사명 기준으로 관심종목을 업데이트 할 때 조건은 충족하지만 결과에 해당 종목이 없는 것을 보고 멘붕을 했던 경험이 있다.<br><br>
#### 알고 보니 회사명이 바뀌어서 같은 종목인데도 다른 종목으로 처리해서 그런 거였다...<br><br>
#### 그런 대참사가 일어났기 때문에 변하지 않는 종목코드도 크롤링 했다.<br><br>
#### 위의 경험으로 mysql의 프라이머리 키가 왜 중요한지 알게 되었다. 수업 시간에 그렇게 들었는데도...<br><br>

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
url = 'https://finance.naver.com/sise/sise_group.naver?type=upjong'
response = requests.get(url)
html = response.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
upjong = []
codes = []
company = []

for i in soup.select("a[href^='/sise/sise_group_detail']"):
    # print(str(i.attrs['href']), i.text)
    link = 'https://finance.naver.com' + str(i.attrs['href'])
    detail_response = requests.get(link)
    detail_html = detail_response.text
    detail_soup = BeautifulSoup(detail_html, 'html.parser')
    for c in detail_soup.select("td > div > a"):
        code = re.findall('(?=)[0-9]+', str(c.attrs['href']))
        print("A" + code[0], i.text, c.text)
        codes.append("A" + code[0])
        upjong.append(str(i.text))
        company.append(str(c.text))
```

```python
df = pd.DataFrame({'code': codes,
                   'company': company,
                   'upjong': upjong})
```

```python
df.to_csv("네이버업종.csv", index = False)
```
